# Exploratory Data Analysis (EDA)

**Dataset:** Iris Dataset
**Objective:**
To explore and understand the structure, distribution, and relationships
between features using visualizations.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

EDA helps to understand the data before applying any machine learning model.
It reveals patterns, trends, and relationships in the dataset.


```python
df = pd.read_csv("iris.csv")
df.head()
```

This step is used to understand the basic structure of the dataset,
including number of rows, columns, and data types.


```python
df.info()
```

The statistical summary provides mean, minimum, maximum,
and spread of numerical features.


```python
df.describe()
```

Histograms are used to visualize the distribution of numerical features.
They show how data values are spread across ranges.


```python
df.hist(figsize=(10,6))
plt.show()
```

The species column is a categorical feature.
Count plot shows the number of samples in each category.

```python
sns.countplot(x="species", data=df)
plt.show()
```

Box plots help in identifying outliers and understanding
the spread of numerical features.

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df.drop("species", axis=1))
plt.show()
```

Correlation heatmap shows the relationship between numerical features.
Higher values indicate strong relationships.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.drop("species", axis=1).corr(), annot=True)
plt.show()
```

Petal length and petal width show strong correlation
and are important features for predicting species.

## Summary

- The dataset is clean and well-structured.
- Numerical features show clear distributions.
- No major outliers are present.
- Petal-related features are important for prediction.
- The dataset is suitable for machine learning models.
